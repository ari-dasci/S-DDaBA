# Dynamic Defense Against Byzantine Poisoning Attacks in Federated Learning

In this notebook we show an example of the behaviour of DDaBA, presented in [this paper](), in Flower. Flower is designed in such a way that you have to initialise a server and add clients to the configuration.

For this reason, the file to initialise the server [server](./server.py) and the file to add clients [client](./client.py) must be launched simultaneously. In order to demonstrate this operation on a notebook, we use the [multiprocess](https://docs.python.org/3/library/multiprocessing.html) library.

```python
#!pip install multiprocess
from multiprocess import Pool
import time
import os

import subprocess

batcmd="dir"
result = subprocess.check_output(batcmd, shell=True)
```

```python
def run_server(cases):
    if cases == 0:
        !python server.py
        time.sleep(3)
    else:
        !python client.py
```

Since server and client outputs overlap, we use only two clients (the minimum required by the framework) to show how it works. We see server initialisation messages, epoch after epoch of client training metrics (overlapping), aggregation, and this repeated for 10 times.

```python
p = Pool(3)
p.map(run_server,[0,1,2])
```
